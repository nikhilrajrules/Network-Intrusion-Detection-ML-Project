```
ar={}
```

```
for i in cols:
  if i!='xAttack':
    l=len(df[i].unique())
    if l>2:
      li=[]
      for j in labels:
        tdf=df[df.xAttack==j]
        x=tdf[i]
        x=np.array(x)
        x=x.flatten()
        fl=np.sum(x)/x.shape[0] if x.shape[0]!=0 else 0
        li.append(fl)
      fl_max=max(li)
      x=df[i]
      x=np.array(x)
      x=x.flatten()
      fl_deno=np.sum(x)/x.shape[0] if x.shape[0]!=0 else 0
      ar_val=fl_max/fl_deno if fl_deno!=0 else 0
    else:
      li=[]
      for j in labels:
        tdf=df[df.xAttack==j]
        x=tdf[i]
        x=np.array(x)
        x=x.flatten()
        su=np.sum(x)
        du=x.shape[0]-su
        fl=su/du if du!=0 else 0
        li.append(fl)
      fl_max=max(li)
      ar_val=fl_max
    ar[i]=ar_val

```
